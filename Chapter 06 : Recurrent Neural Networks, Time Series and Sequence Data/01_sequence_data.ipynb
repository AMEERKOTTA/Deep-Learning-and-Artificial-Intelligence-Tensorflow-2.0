BAG OF WORDS 

+ Suppose you are Doing Document Classification.
+ like Classifying Email to Spam and Non Spam.
+ Then Create a Big Long Feature Vector.
+ One Entry for Each Word in the English Language.
+ Inside the Vector, We insert the Count of how many times each word appeared in the Email.

EXAMPLE OF BAG OF WORDS

+ Suppose there are Only 5 Words in English Language.
+ [insurance, loan, pickles, backpack, football]
+ Lets say the Email Contains `"insurance" 3 times, "loan" 1 time, and other words 0 times`.
+ then the feature vector would be,
  `[3,1,0,0,0]`

PROBLEM WITH BAG OF WORDS

+ When we only count each word, we lose information about each order of the Words.
+ Example "Dog Toy" and "Toy Dog"
+ One is a Toy for Dogs.
+ And the Other is Toy that look like Dog.


WHAT IS A SEQUENCE DATA?

+ Usually we think of a 1D time series Signal.
+ Just like how when we started with Linear Regression, we imagined a line of best fit.
+ Just like with Linear Regression we can Extend the Concept with Multiple Dimensions.
+ For non sequential data, we have NxD matrix.
+ Even if D = 1.

HOW DO WE DEFINE THE SHAPE OF TEH SEQUENCE?

+ How do we represent the Length of the Sequence.
+ Is it N?, or Is it D?
+ Lets use new Letter T as Length of the Sequence.
+ So the Sequence is of Shape (NxTxD)
+ N - number of samples.
+ T - number of steps in the Sequence.
+ D - number of features.

EXAMPLE : LOCATION DATA

+ Suppose we want to build a model to predict the path Employees take to get to work.
+ Record GPS data from their Cars.
+ Here N : One sample would be one person's Single trip to Work.
+ the dataset may contain multiple trips for the same person.
+ D : 2, is (latitude and longitude)
+ T : the number of (lat, long) measurements taken from start to finish the single trip.
+ ie, trip is 30 minutes, cordinates measured every second.
+ then 30 x 60 = 1800.
+ Wouldn't each person take different amount of time to get to work.?
+ In tensorflow/keras we deal only with the equal length sequence.

EXAMPLE : STOCK PRICE

+ A Stock Price is just a Single Value, So D = 1.
+ A Simple Example, Suppose we use a Window of size T = 10, to predict the next value.
+ then N = number of Windows.
+ If we have Sequnce of 100 Stocks.
+ How many windows of length 10 are there.
+ ie, 100 - 10 + 1 = 91
+ ie equation = L - T + 1
+ What if we measure the stock price every day for 500 different stocks.
+ then D = 500.
+ We can still look at the window of length T = 10.
+ Then Single Sample will be of Size TxD = 10x500.
